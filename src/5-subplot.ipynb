### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Data for plots
x = np.arange(0, 11)
y_line = x**2

categories = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 9]
colors_bar = ['red', 'blue', 'green', 'orange']

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# 1. Line plot
axs[0, 0].plot(x, y_line)
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("x")
axs[0, 0].set_ylabel("x squared")
axs[0, 0].grid(True)

# 2. Bar chart
axs[0, 1].bar(categories, values_bar, color=colors_bar)
axs[0, 1].set_title("Bar Chart")

# 3. Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter)
for xi, yi in zip(x_scatter, y_scatter):
    axs[1, 0].annotate(f"({xi},{yi})", (xi, yi), textcoords="offset points", xytext=(5,5), ha='left')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("x")
axs[1, 0].set_ylabel("y")
axs[1, 0].grid(True)

# 4. Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%')
axs[1, 1].set_title("Pie Chart")

plt.tight_layout()
plt.show()
```
